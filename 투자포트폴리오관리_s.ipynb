# gradio를 이용해서 투자전략을 도출하는 앱 만들기


```python
import gradio as gr
import pyupbit
import datetime
import pandas as pd
import mplfinance as mpf
import asyncio
# uvloop 사용 비활성화
asyncio.set_event_loop_policy(None)

# 데이터를 가져오고 분석하는 클래스
class CryptoAnalyzer:
    def __init__(self):
        self.model = "gpt-4-1106-preview"  # 모델 예시

    def get_data(self, ticker, interval, count):
        current_time = datetime.datetime.now()
        formatted_time = current_time.strftime('%Y-%m-%d %H:%M')
        ohlcv = pyupbit.get_ohlcv(ticker=ticker, interval=interval, to=formatted_time, count=count)
        return ohlcv

    def analy_data(self, input_txt):
        response = client.chat.completions.create(
          model="gpt-4-1106-preview",
          response_format={ "type": "json_object" },
              messages=[
                {"role": "system", "content": "You are an exceptionally skilled assistant in technically analyzing the prices of cryptocurrencies. Output the analysis results in JSON format."},
                {"role": "user", "content": f""" {input_txt}코인의 최근 데이터다. 단타투자를 위한 조언을 한글로 제시해줘. 최종파일은 json으로 작성해야한다.
                    """},
              ]
        )

        print(response.choices[0].message.content)
        return response.choices[0].message.content


# 인스턴스 생성
analyzer = CryptoAnalyzer()

# Gradio 인터페이스 설정
def predict():
    data = analyzer.get_data('KRW-BTC', 'minute1', 60)
    analysis_result = analyzer.analy_data(data)
    return analysis_result

iface = gr.Interface(fn=predict, inputs=[], outputs="text", 
                     title="Crypto Investment Prediction",
                     description="Click the button to get the latest 60 minutes investment prediction.")
iface.launch()
```

# 투자 포트폴리오 관리 : 주식, 코인

# 핵심원칙 수립하기

### 1. 다각화 (Diversification)
    - 주식: 다양한 산업 및 지역의 주식을 포함시켜 위험을 분산합니다.
    - 코인: 여러 암호화폐를 포함시켜, 하나의 코인에 의존하는 위험을 줄입니다.   

    
### 2. 리스크 관리 (Risk Management)
    - 주식: 투자 위험을 감안하여, 장기 및 단기 포지션을 적절히 배분합니다.
    - 코인: 고변동성을 고려하여, 전체 포트폴리오에서 코인이 차지하는 비율을 조절합니다.
    
    
### 3. 지속적인 모니터링 및 재평가 (Continuous Monitoring & Rebalancing)
    - 주식과 코인 모두: 시장 변동에 따라 정기적으로 포트폴리오를 재평가하고 필요에 따라 조정합니다.
    
    
### 4. 연구 및 교육 (Research & Education)
    - 주식: 기업의 재무제표, 업계 동향, 경제 지표 등을 연구합니다.
    - 코인: 암호화폐의 기술적 기반, 시장 동향, 규제 변화 등을 주시합니다.
    
    
### 5. 장기적 시각 (Long-Term Perspective)
    - 투자 결정은 단기적인 시장 변동보다는 장기적인 목표와 일치해야 합니다.


### 6. 전문가 상담 (Consulting Experts)
    - 필요시 재정 고문, 투자 전문가의 조언을 구합니다.


### 7. 자신의 감정 관리 (Managing Your Emotions)
    - 시장의 변동성에 휘둘리지 않고, 객관적이고 합리적인 결정을 내리기 위해 노력합니다.
